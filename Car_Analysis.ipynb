```python
import pandas as pd

file_path = 'path_to_your_file.csv'
df = pd.read_csv('./Dataset/cars_ds_final.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
import numpy as np

df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].replace('#REF!', np.nan)
df
```

```python
fuel_type_counts = df['Fuel_Type'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
BodyType = df['Body_Type'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(BodyType, labels=BodyType.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
brand_sales = df['Make'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
brand_sales.plot(kind='bar', color='skyblue')
plt.title('Number of Sales for Each Car Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
top_cars = df.groupby(['Make', 'Model']).size().reset_index(name='Count')

top_10_cars = top_cars.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(12, 8))
plt.barh(top_10_cars['Make'] + ' ' + top_10_cars['Model'], top_10_cars['Count'], color='skyblue')
plt.xlabel('Number of Sales')
plt.ylabel('Car Model')
plt.title('Top 10 Sold Cars')
plt.gca().invert_yaxis()
plt.show()
```
